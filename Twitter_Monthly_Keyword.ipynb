```python
#트위터 크롤링 TXT
import datetime 
from twitterscraper.query import query_tweets 

keyword = 'covid19' 
f = open('c:/data/twitter_5m.txt',mode='w',encoding='utf-8') 
list_of_tweets = query_tweets(keyword, 
                              begindate=datetime.date(2020,5,1), 
                              enddate=datetime.date(2020,5,31))

for tweet in list_of_tweets:
    f.write(str(tweet.text)+'\n')
f.close()
```

```python
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords
tokenizer=RegexpTokenizer('[\w]+')
stop_words=stopwords.words('english')
```

```python
for i in range(6)[1:]:
    f=open(f'c:/data/twitter_5m{i}w.txt', encoding='utf-8')
    lines = f.readlines() 
    f.close()
    result=[]
    for line in lines:
        line=line.lower()
        tokens=tokenizer.tokenize(line)
        stopped_tokens=[i for i in list(tokens) 
                       if not i in stop_words]
        stopped_tokens2=[i for i in stopped_tokens if len(i)>1]
        result.extend(stopped_tokens2)
    if i == 1:
        result1 = result
    elif i == 2:
        result2 = result
    elif i == 3:
        result3 = result
    elif i == 4:
        result4 = result
    elif i == 5:
        result5 = result    
    else:
        pass
```

```python
import datetime 
from twitterscraper.query import query_tweets 

keyword = '코로나' 
f = open('c:/data/twitter/twitter_5m.txt',mode='w',encoding='utf-8') 
list_of_tweets = query_tweets(keyword, 
                              begindate=datetime.date(2020,5,1), 
                              enddate=datetime.date(2020,5,31)) 

for tweet in list_of_tweets: 
    f.write(str(tweet.text)+'\n') 
f.close()
```

```python
from konlpy.tag import Hannanum
import pandas as pd
def flatten(items): #2차원 리스트를 1차원 리스트로 변환
        flatList=[]
        for elem in items:
            if type(elem) == list: #자료형이 리스트이면
                for e in elem: #개별 원소를 추가
                    flatList.append(e)
            else: #리스트가 아니면 그대로 추가
                flatList.append(elem)
        return flatList
```

```python
for i in range(6)[1:]:
    f=open(f'c:/data/twitter/twitter_5m{i}w.txt', encoding='utf-8')
    lines = f.readlines() 
    f.close()
    
    han=Hannanum() #한나눔 패키지
    temp=[]
    for j in range(len(lines)):
        a=lines[j].strip() #공백,줄바꿈,탭 제거
        temp.append(han.nouns(a))     #명사만 추출하여 저장
   
    word_list=flatten(temp)
    word_list=pd.Series([x for x in word_list if len(x)>1])
    
    if i == 1:
        result1 = word_list
    elif i == 2:
        result2 = word_list
    elif i == 3:
        result3 = word_list
    elif i == 4:
        result4 = word_list
    elif i == 5:
        result5 = word_list    
    else:
        pass
```
