```python
from konlpy.tag import Hannanum
import pandas as pd
def flatten(items): #2차원 리스트를 1차원 리스트로 변환
        flatList=[]
        for elem in items:
            if type(elem) == list: #자료형이 리스트이면
                for e in elem: #개별 원소를 추가
                    flatList.append(e)
            else: #리스트가 아니면 그대로 추가
                flatList.append(elem)
        return flatList
```

```python
f=open(f'c:/data/twitter/twitter_5m.txt', encoding='utf-8')
lines = f.readlines() 
f.close()

han=Hannanum() #한나눔 패키지
temp=[]
for j in range(len(lines)):
    a=lines[j].strip() #공백,줄바꿈,탭 제거
    temp.append(han.nouns(a))
```

```python
word_list1 = flatten(temp)
word_list2 = [x for x in word_list1 if len(x)>1]
word_list3 = [i for i in word_list2 if i not in stopwords]
word_list4 = pd.Series(word_list3)
word_list5 = word_list4.value_counts().head(200)
word_list6 = list(word_list5.index)
```

```python
len(word_list6)
```

```python
docs = word_list6
```

```python
docs
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vec=CountVectorizer()
X=vec.fit_transform(docs) 
df2=pd.DataFrame(X.toarray(), columns=vec.get_feature_names())
df2.head()
```

```python
from sklearn.cluster import KMeans
#군집갯수 3개로 설정한 모형
kmeans=KMeans(n_clusters=2, random_state=10).fit(df2)
kmeans.labels_
```

```python
%matplotlib inline
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
#주성분분석 n_components 성분의 갯수
#플로팅을 위하여 3개의 변수를 2개의 주성분으로 축소(차원축소)
pca=PCA(n_components=2, random_state=10)
components=pca.fit_transform(df2)
df3=pd.DataFrame(data=components, columns=['comp 1','comp 2'])
print(df3)
#kmeans.labels_==0
plt.scatter(df3.iloc[kmeans.labels_ == 0,0],
df3.iloc[kmeans.labels_ == 0,1],s=10,c='red',label='cluster 1')

plt.scatter(df3.iloc[kmeans.labels_ == 1,0],
df3.iloc[kmeans.labels_ == 1,1],s=10,c='blue',label='cluster 2')

plt.scatter(df3.iloc[kmeans.labels_ == 2,0],
df3.iloc[kmeans.labels_ == 2,1],s=10,c='green',label='cluster 3')
plt.legend()
```
