```python
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords
tokenizer=RegexpTokenizer('[\w]+') #ÌäπÏàòÎ¨∏Ïûê Ï†úÍ±∞
stop_words=stopwords.words('english') #ÏòÅÏñ¥ Î∂àÏö©Ïñ¥ ÏÇ¨Ï†Ñ
result=[]
for line in lines:
    line=line.lower() #ÏÜåÎ¨∏ÏûêÎ°ú Î∞îÍæ∏Í≥†
    #words.append(line)
    tokens=tokenizer.tokenize(line) #Îã®Ïñ¥ Î¶¨Ïä§Ìä∏Î•º ÎßåÎì§Í≥†
    # Î∂àÏö©Ïñ¥Í∞Ä ÏïÑÎãå Îã®Ïñ¥Îì§
    stopped_tokens=[i for i in list(tokens) 
                   if not i in stop_words]
    # 2Í∏ÄÏûê Ïù¥ÏÉÅÏù∏ Îã®Ïñ¥Îì§
    stopped_tokens2=[i for i in stopped_tokens if len(i)>1]
    result.extend(stopped_tokens2)
```

```python
stopwords=['covid19','post','postcovid19','https','www','la','en','ly','http','com','twitter','de','pic','19','el','para','que','new','2020','bit','covid','','','','','','','','']
result2=[i for i in result if i not in stopwords]
```

```python
%matplotlib inline
from wordcloud import WordCloud
from collections import Counter
import matplotlib.pyplot as plt
#ÏõåÎìú ÌÅ¥ÎùºÏö∞Îìú 
wordcloud=WordCloud(
    font_path='c:/windows/fonts/malgun.ttf',
    width=800, height=800, background_color='white')
#Îã®Ïñ¥Î≥Ñ Ï∂úÌòÑÌöüÏàò
count=Counter(result2)
#print(count)
wordcloud=wordcloud.generate_from_frequencies(count)
fig=plt.figure(figsize=(15,15))
plt.imshow(wordcloud)
plt.show()
```

```python
f=open('c:/data/twitter(Ìè¨Ïä§Ìä∏ÏΩîÎ°úÎÇò).txt',encoding='utf-8')
lines=f.readlines() #ÌååÏùº ÎÇ¥Ïö©ÏùÑ ÏùΩÍ≥†
print(lines[0])
f.close() #ÌååÏùº Îã´Í∏∞
```

```python
from konlpy.tag import Hannanum
han=Hannanum() #ÌïúÎÇòÎàî Ìå®ÌÇ§ÏßÄ
temp=[]
for i in range(len(lines)):
    a=lines[i].strip() #Í≥µÎ∞±,Ï§ÑÎ∞îÍøà,ÌÉ≠ Ï†úÍ±∞
    temp.append(han.nouns(a))     #Î™ÖÏÇ¨Îßå Ï∂îÏ∂úÌïòÏó¨ Ï†ÄÏû•
    
print(temp)  
```

```python
import pandas as pd
def flatten(items): #2Ï∞®Ïõê Î¶¨Ïä§Ìä∏Î•º 1Ï∞®Ïõê Î¶¨Ïä§Ìä∏Î°ú Î≥ÄÌôò
    flatList=[]
    for elem in items:
        if type(elem) == list: #ÏûêÎ£åÌòïÏù¥ Î¶¨Ïä§Ìä∏Ïù¥Î©¥
            for e in elem: #Í∞úÎ≥Ñ ÏõêÏÜåÎ•º Ï∂îÍ∞Ä
                flatList.append(e)
        else: #Î¶¨Ïä§Ìä∏Í∞Ä ÏïÑÎãàÎ©¥ Í∑∏ÎåÄÎ°ú Ï∂îÍ∞Ä
            flatList.append(elem)
    return flatList

word_list=flatten(temp)
#2Í∏ÄÏûê Ïù¥ÏÉÅÏù∏ Îã®Ïñ¥Îì§Îßå Ï∂îÍ∞Ä
word_list=pd.Series([x for x in word_list if len(x)>1])
print(word_list)
```
